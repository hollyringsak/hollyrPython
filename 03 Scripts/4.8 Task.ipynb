# Importing Libraries

```python
# Importing libraries
import pandas as pd
import numpy as np
import os
```

# Importing Data

```python
# Importing dataframe
path = r'/Users/hollyringsak/01-2024 Instacart Basket Analysis'
```

```python
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merge.pkl'))
```

```python
# Grouping by product_name
ords_prods_merge.groupby('product_name')
```

```python
# Getting mean of number_of_orders
ords_prods_merge.groupby('department_id').agg({'number_of_orders': ['mean']})
```

```python
# Mean of number_of_orders for each department is slightly different from the subset because it includes data from the entire dataframe rather than a subset. The numbers are still pretty accurate when using a subset.
```

```python
# Getting the mean of money spent for each loyalty flag
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

```python
# Getting the min/max of money spent for each loyalty flag
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
# Making new column mean_prices to show average amount spent by each user_id
ords_prods_merge['mean_prices'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(100)
```

```python
# Creating a spending flag to mark high and low spenders
ords_prods_merge.loc[ords_prods_merge['mean_prices'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_prices'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Finding median for days_since_last_order
ords_prods_merge.groupby('department_id').agg({'days_since_last_order': ['median']})
```

```python
# Creating new column for the median of days_since_last_order
ords_prods_merge['median_days_since_last_order'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
ords_prods_merge.head(10)
```

```python
# Creating frequency flag
ords_prods_merge.loc[ords_prods_merge['median_days_since_last_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_last_order'] > 10) & (ords_prods_merge['median_days_since_last_order'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_last_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'ords_prods_merge.pkl'))
```
