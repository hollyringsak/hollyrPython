# Create exclusion flag for low-activity customers

```python
# Creating exclusion flag
```

```python
cust_ords.loc[cust_ords['max_order'] <= 5, 'exclusion_flag'] = 'Exclude'
```

```python
cust_ords.loc[cust_ords['max_order'] > 5, 'exclusion_flag'] = 'Safe'
```

```python
# Checking results
```

```python
cust_ords['exclusion_flag'].value_counts(dropna = False)
```

```python
# Creating a copy of the dataframe excluding the rows of data for customers with less than 5 orders
```
